### Temperatures in Montreal averaged over 34 years, 1960â€“1994

```python
import pandas as pd
import skfda
import skfda.representation.basis as basis
```

```python
temp = pd.read_csv("Montreal_temp.csv")
data_matrix = [temp]
grid_points = range(1,366)
Montreal_temp = skfda.FDataGrid(
    data_matrix=data_matrix,
    grid_points=grid_points,
)

basis109 = basis.Fourier(
    n_basis=67,
    domain_range=(0, 365),
)
fd_Montreal_temp = Montreal_temp.to_basis(basis109)
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(9,5),dpi=200)
#Montreal_temp.scatter(fig=fig,s=3,c='red')
fd_Montreal_temp.plot(fig=fig)
fig.axes[0].legend(['Furier','Original'])
```
