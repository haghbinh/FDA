```python
import numpy as np
import skfda
import skfda.representation.basis as basis
```

```python
import numpy as np
t = np.linspace(0, 1, 100)

fd = skfda.FDataGrid(
    data_matrix=[
        np.sin(6 * t**2),  # First function
    ],
    grid_points=t,
)

a=fd.plot()
```

```python
basis1 = basis.Monomial(
    n_basis=4,
    domain_range=(0, 1),
)

basis2 = basis.Fourier(
    n_basis=4,
    domain_range=(0, 1),
)

basis3 = basis.BSpline(
    order=3,
    n_basis=4,
    domain_range=(0, 1),
)
```

```python
fd_basis1 = fd.to_basis(basis1)
fd_basis2 = fd.to_basis(basis2)
fd_basis3 = fd.to_basis(basis3)
```

```python
fig = fd.plot()
fd_basis1.plot(fig=fig)
fd_basis2.plot(fig=fig)
fd_basis3.plot(fig=fig)
fig.axes[0].legend(['Original', 'Monomial', 'Furier','B-Spline'])
```
